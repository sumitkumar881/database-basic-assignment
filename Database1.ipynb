```python
# 1. Create a table called employees with the following structure?
: emp_id (integer, should not be NULL and should be a primary key)
: emp_name (text, should not be NULL)
: age (integer, should have a check constraint to ensure the age is at least 18)
: email (text, should be unique for each employee)
: salary (decimal, with a default value of 30,000).

Write the SQL query to create the above table with all constraints.

Ans:- CREATE TABLE employees (
    emp_id INTEGER NOT NULL PRIMARY KEY,
    emp_name TEXT NOT NULL,
    age INTEGER CHECK (age >= 18),
    email TEXT UNIQUE,
    salary DECIMAL DEFAULT 30000
);
. emp_id INTEGER NOT NULL PRIMARY KEY: The emp_id column is an integer, cannot be
NULL, and is the primary key of the table.

. emp_name TEXT NOT NULL: The emp_name column is a text type and cannot be NULL.

. age INTEGER CHECK (age >= 18): The age column is an integer, and there is a check
constraint ensuring that the age is at least 18.

. email TEXT UNIQUE: The email column is text type and should be unique for each
employee.

. salary DECIMAL DEFAULT 30000: The salary column is a decimal type, with a default
value of 30,000.

# 2. Explain the purpose of constraints and how they help maintain data integrity in a database. Provide examples of common types of constraints.
Ans-Purpose of Constraints in a Database
Constraints in a database are rules enforced on data stored in tables to ensure accuracy, consistency, and reliability.
They help maintain data integrity by preventing invalid or inconsistent data from being entered into the database.
Constraints ensure that the data follows predefined rules and relationships, reducing errors and improving data quality.

How Constraints Help Maintain Data Integrity.
Preventing Invalid Data Entry Constraints restrict what can be inserted or updated, ensuring only valid data is stored.
Ensuring Data Consistency They enforce relationships between tables, preventing orphaned records or mismatched data.
Avoiding Data Duplication Certain constraints prevent duplicate values, ensuring uniqueness.
Maintaining Referential Integrity They help maintain correct relationships between tables.

Common Types of Constraints.

1.Primary Key Constraint
Ensures that each record in a table has a unique identifier.
Example:
CREATE TABLE Employees (
    EmployeeID INT PRIMARY KEY,
    Name VARCHAR(100)
);
Here, EmployeeID must be unique and cannot be NULL.

2.Foreign Key Constraint
Ensures referential integrity by linking a column in one table to the primary key of another table.
Example:
CREATE TABLE Orders (
    OrderID INT PRIMARY KEY,
    EmployeeID INT,
    FOREIGN KEY (EmployeeID) REFERENCES Employees(EmployeeID)
);
Here, EmployeeID in Orders must refer to an existing EmployeeID in the Employees table.
This ensures that EmployeeID in Orders exists in Employees.

3.NOT NULL Constraint
Ensures that a column cannot store NULL values.
Example:

sql
CREATE TABLE Products (
    ProductID INT PRIMARY KEY,
    Name VARCHAR(100) NOT NULL
);
The Name column must always have a value.

4.UNIQUE Constraint
Ensures that all values in a column are distinct.
Example:

sql
CREATE TABLE Customers (
    CustomerID INT PRIMARY KEY,
    Email VARCHAR(255) UNIQUE
);
No two customers can have the same email.

5.CHECK Constraint
Enforces a condition on a column’s values.
Example:

sql
CREATE TABLE Accounts (
    AccountID INT PRIMARY KEY,
    Balance DECIMAL(10,2) CHECK (Balance >= 0)
);
Ensures Balance is never negative.

6.DEFAULT Constraint
Assigns a default value if no value is provided.
Example:

sql
CREATE TABLE Users (
    UserID INT PRIMARY KEY,
    CreatedAt TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);
If no value is inserted for CreatedAt, it defaults to the current timestamp.

# 3.Why would you apply the NOT NULL constraint to a column? Can a primary key contain NULL values? Justify your answer.
Ans:-The NOT NULL constraint is applied to a column to ensure that it always contains a valid value and does not store NULL values. This is important because:
Ensures Data Completeness Prevents missing or incomplete records, which could cause inconsistencies.
Maintains Data Integrity Guarantees that critical fields always contain meaningful values.
Avoids Unexpected Errors  Queries relying on certain columns (e.g., foreign keys or computations) will not fail due to NULL values.
1. Mandatory Fields:
You want to ensure that every row in the table has a value for the column, which is crucial
for certain business operations. For example, in a customer table, the customer_name
column should not allow NULL because every record must have a customer name.

2. Data Integrity:
Applying NOT NULL guarantees that the column cannot be left blank, ensuring the integrity
of the data. For instance, if the column represents an important piece of information (like a
username or email), leaving it NULL could lead to incomplete or incorrect records.

3. Ensuring Consistency:
By disallowing NULL, you ensure that all rows have valid, consistent data for that column,
which is essential for any operations or analysis on that column. For example, in an
employee table, you may want to ensure that all employees have a name and age, as leaving
them NULL would result in incomplete employee records.

4. Simplifying Logic:
The NOT NULL constraint removes the need to check for NULL values in queries, simplifying
the logic when accessing or manipulating data. This is especially helpful in applications or
reporting systems where data completeness is critical.

Can a Primary Key Contain NULL Values?
No, a primary key cannot contain NULL values. Here's why:
1. Uniqueness:
The primary key enforces uniqueness, meaning that each row in the table must have
a unique identifier. If a primary key column allowed NULL, it would violate this
uniqueness requirement, because NULL represents the absence of a value, and
multiple rows with NULL values would technically be considered "equal" (since NULL
is not comparable to any other value, including other NULL values).

2. Identification:
A primary key is used to uniquely identify each record in the table. If it allowed
NULL, there would be no way to uniquely identify the row, making the concept of a
primary key useless. For example, if two rows had a NULL primary key value, they
wouldn't have unique identifiers, which is fundamentally against the purpose of a
primary key.

3. Database Integrity:
Allowing NULL values in the primary key would compromise data integrity, as it
would make it impossible to guarantee that each row in the table could be retrieved
or referenced uniquely. This would lead to inconsistent data relationships, especially
when dealing with foreign keys or joins.

Justification:
To summarize, primary keys cannot contain NULL values because:
Uniqueness:
A primary key must uniquely identify each row, and NULL values cannot be
considered unique.

Identification:
A primary key serves to provide a unique identifier for each record; NULL
cannot act as an identifier.

Data Integrity:
Allowing NULL in primary keys would result in inconsistencies and a lack of
reliability when referencing records, especially when using foreign keys or performing joins.

# 4. Explain the steps and SQL commands used to add or remove constraints on an existing table. Provide an example for both adding and removing a constraint.
Ans:-constraints can be added or removed from an existing table using the ALTER TABLE statement. Below are the steps and SQL commands for both adding and removing constraints.

1. Adding a Constraint.
To add a constraint, use the ALTER TABLE command with the ADD CONSTRAINT clause.

Example: Adding a NOT NULL, UNIQUE, CHECK, and FOREIGN KEY Constraint
Adding a NOT NULL Constraint
You cannot directly add a NOT NULL constraint using ALTER TABLE. Instead, you first update any existing NULL values and then modify the column.

sql
-- Step 1: Ensure no NULL values exist in the column
UPDATE Employees SET Name = 'Unknown' WHERE Name IS NULL;

-- Step 2: Alter the column to NOT NULL
ALTER TABLE Employees MODIFY Name VARCHAR(100) NOT NULL;

. Adding a UNIQUE Constraint
sql
ALTER TABLE Employees ADD CONSTRAINT unique_email UNIQUE (Email);
  Ensures that the Email column does not contain duplicate values.

. Adding a CHECK Constraint
ALTER TABLE Employees ADD CONSTRAINT check_salary CHECK (Salary > 0);
 Ensures that the Salary column always has a positive value.

. Adding a FOREIGN KEY Constraint
ALTER TABLE Orders ADD CONSTRAINT fk_employee FOREIGN KEY (EmployeeID) REFERENCES Employees(EmployeeID);
 Ensures that the EmployeeID in Orders matches an existing EmployeeID in Employees.

2. Removing a Constraint
To remove a constraint, use the ALTER TABLE statement with the DROP CONSTRAINT clause (for most constraints) or MODIFY COLUMN (for NOT NULL).

Example: Removing Constraints
. Removing a UNIQUE Constraint
sql
ALTER TABLE Employees DROP CONSTRAINT unique_email;

. Removing a CHECK Constraint
sql
ALTER TABLE Employees DROP CONSTRAINT check_salary;

.Removing a FOREIGN KEY Constraint
sql
ALTER TABLE Orders DROP CONSTRAINT fk_employee;

.Removing a NOT NULL Constraint
To remove NOT NULL, modify the column to allow NULL values:

sql
ALTER TABLE Employees MODIFY Name VARCHAR(100) NULL;

# 5. Explain the consequences of attempting to insert, update, or delete data in a way that violates constraints. Provide an example of an error message that might occur when violating a constraint.
Ans:-When attempting to INSERT, UPDATE, or DELETE data in a way that violates constraints, the database management system (DBMS) prevents the operation and returns an error message.
Violating constraints ensures data integrity but can cause failures in applications if not properly handled.
Consequences of Violating Constraints During Insert, Update, or Delete Operations
Constraints are rules defined on database columns to ensure the integrity and consistency of the
data. When an attempt is made to insert, update, or delete data that violates any of these
constraints, the database will reject the operation and generate an error message. The
consequences of violating constraints can vary based on the type of constraint being violated, and
the database will enforce data integrity by preventing the operation.
Here’s an overview of what happens when each type of constraint is violated during INSERT,
UPDATE, or DELETE operations:

1. Violating a NOT NULL Constraint.
A NOT NULL constraint ensures that a column cannot contain NULL values. If you attempt to insert
or update a record with a NULL value in a column that has this constraint, the database will reject
the operation.
Example:
 . Inserting Data: If you try to insert a row into a table where a NOT NULL column is left empty:
sql
INSERT INTO employees (emp_id, emp_name, age)
VALUES (1, NULL, 25);

This would violate the NOT NULL constraint on the emp_name column because emp_name cannot
be NULL.

Error Message:
sql
Copy
ERROR: null value in column "emp_name" violates not-null constraint
DETAIL: Failing row contains (1, null, 25).

2. Violating a PRIMARY KEY Constraint
A PRIMARY KEY constraint ensures that each row in a table has a unique identifier and cannot be
NULL. If you attempt to insert or update a record that has a duplicate value in the primary key
column, or if the primary key is set to NULL, the database will reject the operation.
Example:
Inserting Data: If you try to insert a duplicate emp_id value:
sql
INSERT INTO employees (emp_id, emp_name, age)
VALUES (1, 'John Doe', 30);

If there is already a row with emp_id = 1, the insertion will fail because primary keys must be unique.

Error Message:
sql
Copy
ERROR: duplicate key value violates unique constraint "employees_pkey"
DETAIL: Key (emp_id)=(1) already exists.

3. Violating a FOREIGN KEY Constraint
A FOREIGN KEY constraint ensures referential integrity by making sure that a value in one table
corresponds to a valid value in another table. If you attempt to insert, update, or delete a record
that causes a foreign key violation, the database will prevent the operation.

Example:
Inserting Data: If you try to insert an employee with a department_id that does not exist in
the departments table:
sql
Copy
INSERT INTO employees (emp_id, emp_name, department_id)
VALUES (3, 'Jane Smith', 999); -- Department ID 999 doesn't exist.

Error Message:
sql
Copy
ERROR: insert or update on table "employees" violates foreign key constraint "fk_department"
DETAIL: Key (department_id)=(999) is not present in table "departments".

4. Violating a CHECK Constraint
A CHECK constraint ensures that a column value meets a specific condition (e.g., age is greater than
18). If you attempt to insert, update, or delete data that does not meet the condition, the database
will reject the operation.

Example:
Inserting Data: If you attempt to insert an employee with an age less than 18, while a CHECK
constraint enforces that age should be at least 18:
sql
Copy
INSERT INTO employees (emp_id, emp_name, age)
VALUES (4, 'Tommy', 16); -- Age is less than 18

Error Message:
sql
Copy
ERROR: new row for relation "employees" violates check constraint "check_age"
DETAIL: Failing row contains (4, Tommy, 16).

5. Violating a UNIQUE Constraint
A UNIQUE constraint ensures that all values in a column or a combination of columns are distinct. If
you attempt to insert a duplicate value in a column that has a UNIQUE constraint, the database will
reject the operation.

Example:
Inserting Data: If you try to insert a duplicate email address for two employees:
sql
Copy
INSERT INTO employees (emp_id, emp_name, email)
VALUES (5, 'Alice Johnson', 'alice@example.com'),
(6, 'Bob Smith', 'alice@example.com'); -- Duplicate email

Error Message:
sql
Copy
ERROR: duplicate key value violates unique constraint "employees_email_key"
DETAIL: Key (email)=(alice@example.com) already exists.

6. Violating a DEFAULT Constraint
A DEFAULT constraint provides a default value for a column when no value is specified. If a column
has a DEFAULT value, and you explicitly insert a value that contradicts that default value, the
operation will not fail — instead, it will insert the explicitly provided value.

7. Attempting to Delete Rows Referenced by a Foreign Key
If a row in one table is referenced by a FOREIGN KEY constraint in another table, and you attempt to
delete the row in the parent table, it will cause a violation unless the foreign key is defined with ON
DELETE CASCADE or ON DELETE SET NULL.

Example:
Deleting Data: If you attempt to delete a department that has employees in the employees
table without proper ON DELETE CASCADE:
sql
Copy
DELETE FROM departments WHERE department_id = 5;

Error Message (if no ON DELETE CASCADE is set):
sql
Copy
ERROR: update or delete on table "departments" violates foreign key constraint "fk_department"
on table "employees"
DETAIL: Key (department_id)=(5) is still referenced from table "employees".

# Q6.You created a products table without constraints as follows:

CREATE TABLE products (

    REATE TABLE products (

    product_id INT,

    product_name VARCHAR(50),

    price DECIMAL(10, 2));

Now, you realise that?
: The product_id should be a primary keyQ
: The price should have a default value of 50.00
ANS:- To add the required constraints to the existing products table, you need to alter the table and
modify its structure. Specifically, you want to:
1. Set product_id as the primary key.
2. Set a default value of 50.00 for the price column.
Here's how you can achieve this:
Steps and SQL Commands:

1. Add Primary Key to product_id:
To make product_id the primary key, you need to ensure that it is both unique and not null, which is
automatically enforced by the primary key constraint.

2. Set Default Value for price:
You can use the ALTER TABLE statement to set the default value for the price column.
SQL Queries:
sql
Copy
-- Add a primary key constraint to product_id
ALTER TABLE products
ADD CONSTRAINT pk_product_id PRIMARY KEY (product_id);

-- Set the default value for the price column
ALTER TABLE products
ALTER COLUMN price SET DEFAULT 50.00;

Explanation:
1. Add Primary Key:
. ALTER TABLE products ADD CONSTRAINT pk_product_id PRIMARY KEY (product_id);
. This command sets the product_id column as the primary key, which ensures that
each value in the product_id column is unique and not NULL.

2. Set Default Value for price:
. ALTER TABLE products ALTER COLUMN price SET DEFAULT 50.00;
. This command sets the default value of 50.00 for the price column. So, if a new
product record is inserted without specifying a price, it will automatically use 50.00
as the price.

Example Insert:
With these changes, you can insert a product without specifying a price, and it will automatically
have the value 50.00:
sql
Copy
-- Insert product without specifying price (defaults to 50.00)

INSERT INTO products (product_id, product_name)
VALUES (1, 'Product A');

# Q7.You have two tables: Write a query to fetch the student_name and class_name for each student using an INNER JOIN.

To fetch the student_name and class_name for each student using an INNER JOIN, assuming you
have two tables: one for students and another for classes, here is an example SQL query.

Assuming the following table structures:
students table: Contains columns like student_id, student_name, class_id
classes table: Contains columns like class_id, class_name

Query:
sql
Copy
SELECT students.student_name, classes.class_name
FROM students
INNER JOIN classes ON students.class_id = classes.class_id;

Explanation:
We are selecting student_name from the students table and class_name from the classes
table.
The INNER JOIN is used to join the two tables on the common column class_id.
This query will return a list of students and the corresponding class names only for students
who are enrolled in a class (i.e., where there is a match in class_id between the two tables).

# Q8. Consider the following three tables:
Write a query that shows all order_id, customer_name, and product_name, ensuring that all products are
listed even if they are not associated with an order
Hint: (use INNER JOIN and LEFT JOIN).

ANS: Assumed Tables:
1. orders table: Contains columns order_id, customer_id
2. customers table: Contains columns customer_id, customer_name
3. products table: Contains columns product_id, product_name
4. order_products table (or a similar table): Contains order_id, product_id (a many-to-many
relationship table between orders and products)

Query:
sql
Copy
SELECT orders.order_id, customers.customer_name, products.product_name
FROM products
LEFT JOIN order_products ON products.product_id = order_products.product_id
LEFT JOIN orders ON order_products.order_id = orders.order_id
LEFT JOIN customers ON orders.customer_id = customers.customer_id;

Explanation:
1. LEFT JOIN products and order_products: We start with the products table to ensure that all
products are included, even if they have no corresponding order.
2. LEFT JOIN orders: This ensures that all orders are included, even if a product is not
associated with them.
3. LEFT JOIN customers: To get the customer_name for each order, we join the customers
table, using the customer_id from the orders table.

# Q9.Given the following tables:
Write a query to find the total sales amount for each product using an INNER JOIN and the SUM() function.

ANS: Assumed Tables:
1. products table: Contains columns like product_id, product_name, etc.
2. order_products table (or similar): Contains columns like order_id, product_id, quantity,
price (assuming quantity and price are needed to calculate the sales amount).
3. orders table: Contains columns like order_id, order_date, etc.

In this case, we'll assume that order_products contains the quantity and price for each product in an
order, and we'll compute the total sales by multiplying quantity by price.

Query:
sql
Copy
SELECT products.product_name, SUM(order_products.quantity * order_products.price) AS
total_sales
FROM products
INNER JOIN order_products ON products.product_id = order_products.product_id
GROUP BY products.product_name;

Explanation:
1. INNER JOIN between products and order_products: This join ensures we only include
products that have sales (i.e., products that are in the order_products table).
2. SUM(order_products.quantity * order_products.price): The SUM() function is used to
calculate the total sales for each product by multiplying the quantity of each product sold by
its price and then summing them.
3. GROUP BY products.product_name: This ensures the result is grouped by each product, so
you get the total sales for each product.

# Q10. You are given three tables:
Write a query to display the order_id, customer_name, and the quantity of products ordered by each
customer using an INNER JOIN between all three tables

ANS: Assumed Tables:
1. orders table: Contains columns like order_id, customer_id
2. customers table: Contains columns like customer_id, customer_name
3. order_products table: Contains columns like order_id, product_id, quantity

Query:
sql
Copy
SELECT orders.order_id, customers.customer_name, SUM(order_products.quantity) AS
total_quantity
FROM orders
INNER JOIN customers ON orders.customer_id = customers.customer_id
INNER JOIN order_products ON orders.order_id = order_products.order_id
GROUP BY orders.order_id, customers.customer_name;

Explanation:
1. INNER JOIN orders and customers: This join connects the orders table with the customers
table on the customer_id, ensuring that we have the customer_name for each order.
2. INNER JOIN order_products: This join connects the orders table with the order_products
table on the order_id, ensuring that we can access the quantity of products ordered.
3. SUM(order_products.quantity): The SUM() function is used to calculate the total quantity of
products ordered for each order (in case multiple products are ordered in a single order).
4. GROUP BY orders.order_id, customers.customer_name: This groups the result by order_id
and customer_name so that you get the quantity of products ordered by each customer for
each order.
```

```python
# Q1.First Normal Form (1NF):
 a. Identify a table in the Sakila database that violates 1NF. Explain how you would normalize it to achieve 1NF.

Ans:-First Normal Form (1NF) Overview:
In First Normal Form (1NF), a table is considered normalized if:
    All columns contain atomic (indivisible) values. There should be no repeating groups or
      arrays in a single column.
   Each record (row) is unique, and there is a primary key to uniquely identify each record.
   The order in which the data is stored does not matter.
     Table in Sakila Database That Violates 1NF:
    One table in the Sakila database that could potentially violate 1NF is the rental table. If there were a
    scenario where the inventory_id or rental_date column contained multiple values (e.g., lists of
    items or multiple rental dates for a single rental transaction), it would violate 1NF.

Example of a Violating Rental Table:
Suppose we had the following representation (which is a violation of 1NF):
rental_id       rental_date           inventory_id customer_id
1            2023-03-01,2023-03-02 1, 2      100
2            2023-03-03 3                    101
3           2023-03-04 4                     102

. Violation: The rental_date and inventory_id columns contain multiple values (non-atomic
data) separated by commas.
How to Normalize the Table to Achieve 1NF:
To bring this table into First Normal Form (1NF), you must ensure that:

 Each column contains only atomic (indivisible) values.
 There are no repeating groups or multiple values in a single column.

Normalized Rental Table (1NF):
You can split the multi-valued columns into individual rows, ensuring that each column contains only
one value per row.

rental_id     rental_date             inventory_id customer_id
1          2023-03-01 1                     100
1          2023-03-02 2                     100
2          2023-03-03 3                     101
3         2023-03-04 4                      102

 Now, each rental_date and inventory_id is atomic (contains a single value).
 The rental_id is repeated for each rental record, but the data is now properly normalized.
Key Steps in Normalization:
1. Remove repeating groups: Each column should contain only a single value per record.
2. Ensure atomicity: Columns like rental_date and inventory_id must not contain lists or
multiple values in a single row.
3. Ensure unique records: Each row should have a unique identifier, which is typically ensured
by the primary key.

# Q2.Second Normal Form (2NF):
 a. Choose a table in Sakila and describe how you would determine whether it is in 2NF. If it violates 2NF, explain the steps to normalize it.

Ans:-Second Normal Form (2NF) Overview:
A table is in Second Normal Form (2NF) if:
1. It is in First Normal Form (1NF) (i.e., it has no repeating groups, and all columns contain
atomic values).
2. It eliminates partial dependencies, meaning that all non-key attributes must depend on the
entire primary key and not just part of it. This is specifically relevant to composite primary
keys, where the primary key consists of more than one column.

Key Concept of 2NF:
 Partial dependency exists when a non-key column is dependent on only part of a composite
primary key, instead of the whole key.
 If there are no composite primary keys, a table will automatically satisfy 2NF (since there is
no "part of the key" to have dependencies on).

Table in Sakila to Examine for 2NF:
Let's consider the rental table in the Sakila database for analysis.
Schema of the rental table (simplified):
 rental_id (Primary Key)
 rental_date
 inventory_id
 customer_id
 staff_id
 return_date
 amount (amount spent on this rental, perhaps in payment table, for simplicity)
We can consider a composite primary key case with a rental_id and inventory_id in certain contexts
(like in the inventory table).

Step 1: Check if the table is in 1NF
We first check if the table is in 1NF:

 Atomic columns: All columns contain atomic values.
 Unique records: Each record is unique, often achieved by a primary key.
Assume the rental table is already in 1NF (no repeating groups, all columns contain atomic values).

Step 2: Check for Partial Dependencies
Now, to check if the table is in 2NF, we need to examine if there are partial dependencies in the
table. This can occur if the primary key is composite (consists of multiple columns), and non-key
attributes depend on only part of the primary key.

For example, if the primary key of the rental table was (rental_id, inventory_id), we would check if
any non-key columns are dependent on just one part of this composite key.
Let's examine possible partial dependencies:

 rental_date and return_date are likely fully dependent on rental_id (the primary key for
that rental).
 inventory_id is part of the primary key and should not be dependent on anything else.
 customer_id is related to the rental_id and should not be partially dependent on
inventory_id.

If we find that, for example, amount (if stored in the rental table) depends on inventory_id rather
than the full rental_id, this is a partial dependency, and the table is not in 2NF.

Step 3: Normalize to 2NF
If we identify partial dependencies, we will split the table to eliminate them.
Example of Partial Dependency:
Assume that amount (payment amount) is stored in the rental table but depends on inventory_id
(since the amount may vary based on the movie rented). This would be a partial dependency.

Steps to Normalize to 2NF:
1. Remove Partial Dependency: We need to move the attributes that depend on part of the
composite key to a new table.
2. Create a New Table for Amount: Since amount depends on inventory_id (which is part of
the composite primary key), it should be moved to a new table, where inventory_id is the
primary key.

Create a new table, e.g., inventory_amount, to store the amount based on inventory_id:
sql
CREATE TABLE inventory_amount (
inventory_id INT PRIMARY KEY,
amount DECIMAL(10, 2) NOT NULL
);

3. Link Tables: Modify the original rental table to exclude amount and establish a relationship
with the inventory_amount table.

sql
CREATE TABLE rental (
rental_id INT PRIMARY KEY,
rental_date DATE,
return_date DATE,
customer_id INT,
staff_id INT,
inventory_id INT,
FOREIGN KEY (inventory_id) REFERENCES inventory(inventory_id),
FOREIGN KEY (customer_id) REFERENCES customer(customer_id),
FOREIGN KEY (staff_id) REFERENCES staff(staff_id)
);

4. Adjust the Relationship: To retrieve the total amount for a rental, you would now join the
rental and inventory_amount tables based on inventory_id.

# Q3.Third Normal Form (3NF)
 a. Identify a table in Sakila that violates 3NF. Describe the transitive dependencies present and outline the steps to normalize the table to 3NF.

Ans:-Third Normal Form (3NF) Overview:
A table is in Third Normal Form (3NF) if:
1. It is in Second Normal Form (2NF) (i.e., it is in 1NF and has no partial dependencies).
2. There are no transitive dependencies — this means that non-key attributes should not
depend on other non-key attributes.

In simpler terms, transitive dependencies occur when:
 A non-key column depends on another non-key column, which is also dependent on the
primary key.

To bring a table to 3NF, any transitive dependency must be removed, typically by creating new
tables and establishing relationships between them.
Table in Sakila Database that Might Violate 3NF:
A good example of a table that could violate 3NF in the Sakila database is the customer table.

Schema of the customer table (simplified):
 customer_id (Primary Key)
 first_name
 last_name
 email
 address_id (Foreign Key referencing address table)
 store_id (Foreign Key referencing store table)
 active (Indicates whether the customer is active)
 create_date
 last_update
 address (non-key column: Street, City, Zipcode stored in the address table).

Step 1: Identify if the Table is in 2NF
For 3NF normalization, the table must already be in 2NF. Based on the schema above, we assume:
 1NF: The customer table is in 1NF because all columns contain atomic values.
 2NF: Since customer_id is the primary key, and there are no partial dependencies in the
table, it is in 2NF.

Step 2: Identify Transitive Dependencies
Now, let's look for transitive dependencies. A transitive dependency occurs when one non-key
attribute depends on another non-key attribute, which is also dependent on the primary key.
Possible transitive dependencies in the customer table could be:

1. address: The address column in the customer table may include multiple pieces of
information like street, city, and zipcode, which depend on address_id, but address_id itself
depends on the customer_id (since it’s a foreign key to the address table).
2. city: If we store city directly in the customer table (instead of only referencing it from the
address table), then city would be dependent on address_id, which is a transitive
dependency (since address_id depends on customer_id).

Step 3: Normalize to 3NF
To remove the transitive dependency, we can break the table into smaller tables where each table
has a single non-key dependency and no non-key attributes depend on another non-key attribute.

New Normalized Table Structure
1. Updated customer Table (3NF Compliant)
sql
CREATE TABLE customer (
    customer_id INT PRIMARY KEY,
    first_name VARCHAR(50),
    last_name VARCHAR(50),
    email VARCHAR(100),
    address_id INT,
    FOREIGN KEY (address_id) REFERENCES address(address_id)
);

2. New address Table
sql
CREATE TABLE address (
    address_id INT PRIMARY KEY,
    address VARCHAR(255),
    city_id INT,
    postal_code VARCHAR(10),
    FOREIGN KEY (city_id) REFERENCES city(city_id)
);

3. New city Table
sql
CREATE TABLE city (
    city_id INT PRIMARY KEY,
    city VARCHAR(100),
    country_id INT,
    FOREIGN KEY (country_id) REFERENCES country(country_id)
);

4. New country Table

REATE TABLE country (
    country_id INT PRIMARY KEY,
    country VARCHAR(100)
);

# Q4. Normalization Process:
 a. Take a specific table in Sakila and guide through the process of normalizing it from the initial unnormalized form up to at least 2NF.

Ans:-Normalization Process:
Normalization is the process of organizing data in a database to minimize redundancy and improve
data integrity. The process involves breaking a table into smaller tables and establishing
relationships between them. We will guide you through normalizing a specific table from the Sakila
database from an Unnormalized Form (UNF) up to Second Normal Form (2NF).
Table Selection: rental table (example).

Let's start with the rental table as an example. Below is a simplified version of the rental table,
assuming it is initially in an Unnormalized Form (UNF).
Example: Unnormalized rental Table
rental_id    	rental_date	      inventory_id	         customer_id	        return_date      	staff_id    	store_id	  store_address	 store_city	  store_country
1	          2024-01-10	           100                  	5	                2024-01-15	         2	            1	        123 Main St   New York         	USA
2	          2024-01-12	           101	                  6                	2024-01-16         	3	              1	        123 Main St   New York	        USA
3         	2024-01-15	           102	                  5                	2024-01-18	        2              	2        	456 Oak Ave	  Los Angeles     	USA

Step 1: 1st Normal Form (1NF)
To be in First Normal Form (1NF), the table should meet the following criteria:
 All columns contain atomic values (i.e., no repeating groups or arrays).
 Each record in the table is unique, which is typically enforced by a primary key.

Unnormalized Form (UNF) Issues:
 The customer_name, address, and city columns contain repeating information for each
rental. This is a violation of 1NF since these columns should be stored separately.
 There may also be data redundancy in terms of customer-related information across
multiple rentals.

Normalizing to 1NF:
1. We remove the repeating groups (e.g., customer-related data) and make sure all columns
contain atomic values.
2. We introduce a primary key (rental_id) to ensure that each record is unique.
The rental table is now in 1NF:

Step 2: 2nd Normal Form (2NF)
To be in Second Normal Form (2NF), the table must:
1. Be in 1NF.
2. Have no partial dependencies — all non-key attributes must depend on the entire primary
key (in case of a composite key).

Identifying Partial Dependencies:
 Primary Key: In the rental table, rental_id is the primary key.
 Partial dependency: We check if any non-key columns depend on only part of the composite
key (if we had one). Since our primary key is rental_id, there is no composite key here, so
partial dependency isn't an issue.

However, there are transitive dependencies that violate 2NF:
 Customer details (like customer_name, address, and city) depend on customer_id, but
customer_id is not part of the primary key. Therefore, these attributes are transitively
dependent on rental_id through customer_id.

Steps to Normalize to 2NF:
1. Create a separate customer table to store customer details.
2. Create a separate address table to store the address-related data.
3. Remove customer and address-related data from the rental table and replace them with
customer_id and address_id.

Normalized rental Table (1NF)
rental_id	        rental_date	          inventory_id        	customer_id	       return_date      	staff_id      	store_id
1               	2024-01-10             	100	                 5	               2024-01-15           	2	              1
2	              2024-01-12	             101	                 6                	2024-01-16           	3             	1
3	              2024-01-15	             102	                 5	               2024-01-18	            2	              2

store_id         	store_address         	store_city         	store_country
1              	123 Main St	             New York              	USA
2             	456 Oak Ave	            Los Angeles            	USA

 We removed the transitive dependency between customer_name, address, and city by
creating separate customer and address tables.
 The rental table now only contains rental_id, inventory_id, customer_id, payment_amount,
and rental_duration, with customer_id linking to the customer table and address_id linking.

# Q5.CTE Basics:
 a. Write a query using a CTE to retrieve the distinct list of actor names and the number of films they have acted in from the actor and film_actor
Ans:-To retrieve the distinct list of actor names and the number of films they have acted in using a
Common Table Expression (CTE), you can follow this approach.

Example:
Query:
sql
Copy
WITH ActorFilmCount AS (
SELECT
a.actor_id,
CONCAT(a.first_name, ' ', a.last_name) AS actor_name,
COUNT(fa.film_id) AS film_count
FROM actor a
JOIN film_actor fa ON a.actor_id = fa.actor_id

GROUP BY a.actor_id, actor_name
)
SELECT * FROM ActorFilmCount;

Explanation:
 CTE Definition (ActorFilmCount):
o We select actor_id, concatenate first_name and last_name to get the full name of
the actor as actor_name.
o We join the actor table with the film_actor table using the actor_id to get the films
each actor has acted in.
o We use COUNT(fa.film_id) to count the number of films each actor has appeared in.
o GROUP BY is used to group by actor_id and actor_name to count the films per
actor.

 Final SELECT: The final SELECT statement retrieves the data from the CTE ActorFilmCount,
which contains the actor_name and their corresponding film_count.
To create a Common Table Expression (CTE) that combines information from the film and language
tables to display the film title, language name, and rental rate, you can follow this query structure:

Query:
sql
Copy
WITH FilmLanguageInfo AS (
SELECT
f.title AS film_title,
l.name AS language_name,
f.rental_rate
FROM film f
JOIN language l ON f.language_id = l.language_id
)
SELECT * FROM FilmLanguageInfo;

Explanation:
 CTE Definition (FilmLanguageInfo):
o We select the title from the film table and alias it as film_title.
o We select the name from the language table and alias it as language_name.
o We also select the rental_rate from the film table.
o We join the film table with the language table using language_id to get the correct
language for each film.

 Final SELECT: The final SELECT statement retrieves the film title, language name, and rental
rate from the CTE FilmLanguageInfo.

# Q6.CTE with Joins:
 a. Create a CTE that combines information from the film and language tables to display the film title,
Ans:-
language name,
CTE with Joins: Combining Film and Language Information
Problem Statement
Create a Common Table Expression (CTE) that retrieves film titles along with their language names from the film and language tables.

Solution Using CTE
sql
Copy
Edit
WITH FilmLanguage AS (
    SELECT
        f.film_id,
        f.title AS film_title,
        l.name AS language_name
    FROM film f
    JOIN language l ON f.language_id = l.language_id
)
SELECT * FROM FilmLanguage
ORDER BY language_name, film_title;

Explanation
1.CTE (FilmLanguage)
Selects film_id, title (renamed as film_title), and name (renamed as language_name).
Joins film with language on language_id.

2.Main Query
Retrieves all columns from the CTE.
Orders the results by language_name first, then by film_title.

# Q7. CTE for Aggregation:
a. Write a query using a CTE to find the total revenue generated by each customer (sum of payments) from the customer and payment tables.

Ans:-To find the total revenue generated by each customer, which is the sum of all payments made by
each customer, we can use a Common Table Expression (CTE). Here's the query that retrieves the
total revenue for each customer from the customer and payment tables:

Query:
sql
Copy
WITH CustomerRevenue AS (
SELECT
c.customer_id,
CONCAT(c.first_name, ' ', c.last_name) AS customer_name,
SUM(p.amount) AS total_revenue
FROM customer c
JOIN payment p ON c.customer_id = p.customer_id
GROUP BY c.customer_id, customer_name
)
SELECT * FROM CustomerRevenue;

Explanation:
 CTE Definition (CustomerRevenue):
o customer_id: We select the customer_id from the customer table.
o customer_name: We concatenate first_name and last_name from the customer
table to create a full name for each customer.
o SUM(p.amount): We sum the amount column from the payment table to calculate
the total revenue (payments made) for each customer.
o JOIN: We join the customer table with the payment table on customer_id to link
each customer with their payments.
o GROUP BY: We group the results by customer_id and customer_name to aggregate
the payment amounts for each customer.
 Final SELECT: The final SELECT retrieves the customer_id, customer_name, and the
total_revenue from the CTE CustomerRevenue.

# Q8.CTE with Window Functions:
a. Utilize a CTE with a window function to rank films based on their rental duration from the film table.

Ans:-To rank films based on their rental duration using a Common Table Expression (CTE) with a window
function, you can use the RANK() function to assign a rank to each film based on the
rental_duration. This will allow you to see which films have the highest rental duration.

Here's the query:
Query:
sql
Copy
WITH FilmRank AS (

SELECT
f.title AS film_title,
f.rental_duration,
RANK() OVER (ORDER BY f.rental_duration DESC) AS rental_rank
FROM film f
)
SELECT * FROM FilmRank;

Explanation:
 CTE Definition (FilmRank):
o film_title: Select the title from the film table to get the name of each film.
o rental_duration: Select the rental_duration from the film table to rank films based
on how long they are rented.
o RANK() OVER (ORDER BY f.rental_duration DESC): This window function assigns a
rank to each film, ordered by rental_duration in descending order. Films with the
longest rental duration will be ranked first (rank 1).
 RANK() is a window function that assigns a rank to each row within the
partition of a result set. In this case, we are not using any partitioning, so it
ranks all the rows across the entire table.
 The ORDER BY f.rental_duration DESC orders the films by rental_duration
in descending order, so the highest rental duration gets the highest rank.
 Final SELECT: The final SELECT retrieves the film_title, rental_duration, and rental_rank
from the CTE FilmRank.

# Q9.CTE and Filtering:
 a. Create a CTE to list customers who have made more than two rentals, and then join this CTE with the customer table to retrieve additional customer details.

Ans:-To list customers who have made more than two rentals, we will first create a Common Table
Expression (CTE) that aggregates the number of rentals for each customer. Then, we will join this
CTE with the customer table to retrieve additional customer details.

WITH FrequentRenters AS (
    SELECT
        customer_id,
        COUNT(rental_id) AS rental_count
    FROM rental
    GROUP BY customer_id
    HAVING COUNT(rental_id) > 2
)
SELECT
    c.customer_id,
    c.first_name,
    c.last_name,
    c.email,
    c.active,
    f.rental_count
FROM FrequentRenters f
JOIN customer c ON f.customer_id = c.customer_id
ORDER BY f.rental_count DESC;

Explanation
1.CTE (FrequentRenters)
Groups the rental table by customer_id.
Counts the number of rentals per customer.
Filters to include only customers with more than two rentals (HAVING COUNT(rental_id) > 2).

2.Main Query
Retrieves customer details (first_name, last_name, email, active status) from the customer table.
Joins with the FrequentRenters CTE on customer_id.
Orders results by rental count (descending) to show the most frequent renters first.

# Q10.CTE for Date Calculations:
a. Write a query using a CTE to find the total number of rentals made each month, considering the rental_date from the rental table

Ans:-To find the total number of rentals made each month, we can use a Common Table Expression
(CTE) to group the rental_date by month and year and then calculate the total number of rentals for
each month.
Here's the query:

Query:
sql
Copy
WITH MonthlyRentals AS (
SELECT
YEAR(r.rental_date) AS rental_year,
MONTH(r.rental_date) AS rental_month,

COUNT(r.rental_id) AS total_rentals
FROM rental r
GROUP BY YEAR(r.rental_date), MONTH(r.rental_date)
)
SELECT * FROM MonthlyRentals
ORDER BY rental_year DESC, rental_month DESC;

Explanation:
1. CTE Definition (MonthlyRentals):
o YEAR(r.rental_date): Extracts the year from the rental_date.
o MONTH(r.rental_date): Extracts the month from the rental_date.
o COUNT(r.rental_id): Counts the total number of rentals for each month.
o GROUP BY YEAR(r.rental_date), MONTH(r.rental_date): Groups the results by year
and month to aggregate the rental counts for each month.

2. Final SELECT:
o The final SELECT retrieves the year, month, and total number of rentals for each
month from the CTE MonthlyRentals.
o ORDER BY rental_year DESC, rental_month DESC: Orders the results by year and
month in descending order to show the most recent months first.

# Q11.CTE and Self-Join:
 a. Create a CTE to generate a report showing pairs of actors who have appeared in the same film together, using the film_actor table.

Ans:-To generate a report showing pairs of actors who have appeared in the same film together using the
film_actor table, we can create a Common Table Expression (CTE) that performs a self-join on the
film_actor table. This self-join will allow us to find pairs of actors who have appeared in the same
film.

WITH ActorPairs AS (
    SELECT
        fa1.film_id,
        fa1.actor_id AS actor1_id,
        fa2.actor_id AS actor2_id
    FROM film_actor fa1
    JOIN film_actor fa2
        ON fa1.film_id = fa2.film_id
        AND fa1.actor_id < fa2.actor_id  -- Avoid duplicate pairs & self-pairs
)
SELECT
    a1.first_name || ' ' || a1.last_name AS actor1_name,
    a2.first_name || ' ' || a2.last_name AS actor2_name,
    f.title AS film_title
FROM ActorPairs ap
JOIN actor a1 ON ap.actor1_id = a1.actor_id
JOIN actor a2 ON ap.actor2_id = a2.actor_id
JOIN film f ON ap.film_id = f.film_id
ORDER BY f.title, actor1_name, actor2_name;

Explanation
1.CTE (ActorPairs)
Selects pairs of actors (actor1_id and actor2_id) who appeared in the same film (film_id).
Self-joins film_actor (fa1 and fa2) on film_id to match actors in the same film.
Uses fa1.actor_id < fa2.actor_id to avoid duplicate pairs (e.g., (1,2) instead of both (1,2) and (2,1)).

2.Main Query
Joins ActorPairs with the actor table (twice) to get actor names.
Joins film to retrieve the film title.
Orders results by film title and actor names

# Q12.CTE for Recursive Search:
 a. Implement a recursive CTE to find all employees in the staff table who report to a specific manager, considering the reports_to column.table

Ans:-To implement a recursive Common Table Expression (CTE) that finds all employees who report to a
specific manager in the staff table, we can use the reports_to column. The reports_to column in the
staff table typically refers to the staff_id of the manager for each employee.
The recursive CTE will allow us to find all employees who report directly or indirectly to a specific
manager, considering the hierarchical reporting structure.

Structure:
1. Base Case: The base case will select the employees who directly report to a specific
manager.
2. Recursive Case: The recursive part will continue to find employees who report to the
employees selected in the previous step.

WITH RECURSIVE EmployeeHierarchy AS (
    -- Base Case: Find the manager
    SELECT
        staff_id,
        first_name,
        last_name,
        reports_to,
        1 AS level
    FROM staff
    WHERE staff_id = 1  -- Change to the specific manager's ID

    UNION ALL

    -- Recursive Case: Find employees reporting to the above employees
    SELECT
        s.staff_id,
        s.first_name,
        s.last_name,
        s.reports_to,
        eh.level + 1
    FROM staff s
    JOIN EmployeeHierarchy eh ON s.reports_to = eh.staff_id
)
SELECT * FROM EmployeeHierarchy
ORDER BY level, staff_id;

Explanation:
1. Base Case:
o  The base case in the recursive CTE selects employees who directly report to a
   manager with a specific staff_id (in this example, manager with staff_id = 1).
o  WHERE s.reports_to = 1: This filters the employees who directly report to the
   manager (replace 1 with the manager’s actual staff_id).

2. Recursive Case:
o The recursive part of the CTE continues to select employees who report to the
   employees selected in the previous step.
o The JOIN is done between the staff table and the EmployeeHierarchy CTE itself on
  the reports_to column to find employees who report to the staff members
  identified in the previous step.

3. Final SELECT:
o  The final SELECT retrieves all the rows from the EmployeeHierarchy CTE, which will
   contain all employees who directly or indirectly report to the specified manager.
```
